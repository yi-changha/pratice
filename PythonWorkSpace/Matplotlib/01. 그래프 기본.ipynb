# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as pit
```

# 1. 그래프 기본

```python
x = [1,2,3]
y = [2,4,8]
pit.plot(x,y)
```

```python
print(pit.plot(x,y))
```

```python
pit.plot(x,y)
pit.show()  # 불필요한 text를 제거
```

## Title 설정

```python
pit.plot(x,y)
pit.title('Line Graph')
```

```python
pit.plot(x,y)
pit.title('꺾은선 그래프')
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # WIndows 맑은고딕
# matplotlib.rcParms['font.family']= 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# 한글 폰트 사용 시, 마그너스 글자가 깨지는 현상을 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 이름만 간단하게 확인
```
