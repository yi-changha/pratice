# 5. 파일 저장

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # WIndows 맑은고딕
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# 한글 폰트 사용 시, 마그너스 글자가 깨지는 현상을 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)

plt.savefig('graph.png', dpi=100) # 파일로 저장, 해상도 조정
```

```python
plt.figure(dpi=50) # 화면에 보여지는 해상도
plt.plot(x,y)
plt.savefig('graph_50.png', dpi=100) # 저장되는 해상도
#savefig 함수에서 dpi값 미설정 시, 기본값(figure 함수값)에 따라서 저장됨.
```
