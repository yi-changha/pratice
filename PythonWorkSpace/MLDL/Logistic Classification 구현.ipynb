# Logistic Classification

```python
import tensorflow as tf

x_data = [[1, 2],
          [2, 3],
          [3, 1],
          [4, 3],
          [5, 3],
          [6, 2]]
y_data = [[0],
          [0],
          [0],
          [1],
          [1],
          [1]]
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=2))
# use sigmoid activation for 0~1 problem
tf.model.add(tf.keras.layers.Activation('sigmoid'))
```

```python
tf.model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])
tf.model.summary()
```

```python
history = tf.model.fit(x_data, y_data, epochs=200)
```

```python
print("Accuracy: ", history.history['accuracy'][-1])
```

 ### loss function == 'binary_crossentropy' 보다 정확도 높은 결과를 위해 loss == 'mse' 이용

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.RMSprop(), metrics=['accuracy'])
tf.model.summary()
history = tf.model.fit(x_data, y_data, epochs=200)
```

```python
print("Accuracy: ", history.history['accuracy'][-1])
```

### epoch이 낮을 때 accuracy가 개선되는 것을 확인 가능함.
