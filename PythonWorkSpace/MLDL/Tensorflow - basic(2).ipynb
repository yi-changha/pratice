# Tensorflow - basic

```python
import tensorflow as tf
```

```python
키 = 170 #키
신발 = 260 #신발 사이즈
#신발 = 키*a+b

a = tf.Variable(0.1) # 초기값 설정
b = tf.Variable(0.2)

def 손실함수():
    예측값= 키 * a + b
    return tf.square(260 - 예측값) #손실값 = 실제값 - 예측값


opt = tf.keras.optimizers.Adam(learning_rate=0.1) 
# 경사하강법 / Adam 알아서 경사를 바꿔줌

for i in range(300):
    opt.minimize(손실함수, var_list=[a,b]) # var_list 변화를 줄 list
    print(a.numpy(),b.numpy()) # .numpy 실제값만 출력
```

# 대학원 입학 시뮬레이터

## 1. 딥러닝 model 디자인하기

```python
# Sequential 신경망 레이어 생성
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='tanh'),   
    # hidden layer / tf.keras.layers.Dense(노드의 개수, activation function)
    tf.keras.layers.Dense(128, activation='tanh'),  # 관습적으로 노드의 개수 2의 배수
    tf.keras.layers.Dense(1, activation='sigmoid'),    
    # 마지막으로 예측하는 노드의 개수  0~1사이의 결과 : sigmoid
])
```

## 2. model compile 하기

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# optimizer 기울기 조정, loss함수=binary_croosentropy 0과1로 나눔
```

## 3. model 학습(fit) 시키기

```python
import pandas as pd
```

```python
data = pd.read_csv('gpascore.csv')
data
```

### * 데이터 전처리 / pandas 연습

```python
data.isnull().sum() #빈칸 개수 세기
```

```python
data = data.dropna()
data.isnull().sum()
# data.fillna(100) # 빈칸채우기
```

```python
data['gpa']
```

```python
data['gre'].max()
```

```python
data['gre'].count() # gre에 속한 데이터 개수
```

```python
y데이터 = data['admit'].values
x데이터 = []
for i, rows in data.iterrows():  # data라는 데이터 프레임을 가로 한줄씩 출력
    x데이터.append([rows['gre'], rows['gpa'], rows['rank']]) # X데이터에 추가

x데이터
```

```python
import numpy as np
```

```python
model.fit(np.array(x데이터), np.array(y데이터), epochs=1000)
# model.fit( x데이터, y데이터, epochs=200)
# x에는 학습데이터, y는 실제정답, epochs 얼마나 학습시킬 것인지.
# list를 그대로 넣을 수 없고 numpy array나 tf tensor 이용
```
