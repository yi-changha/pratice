# Linear Regression (2)

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np

tf.__version__
```

## 1. 데이터셋 생성

```python
x_data = np.array([1, 2, 3, 4, 5, 6])
y_data = np.array([3, 4, 5, 6, 7, 8])  
```

```python
x_data.shape
```

```python
y_data.shape
```

## 2. 모델(model) 구축

```python
model = Sequential() #모델
model.add(Flatten(input_shape=(1,)))       # 입력층
model.add(Dense(1, activation='linear'))   # 출력층

# model.add(Dense(1, input_shape=(1,), activation='linear'))  # 입력층 + 출력층
```

## 3. 모델 컴파일 및 summary

```python
# [3] 모델 (model) 컴파일 및 summary

model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')
# SGD : 학습 알고리즘
# learning_rate : 학습률
# loss='mse : 손실함수

model.summary()
```

## 4. 모델 학습

```python
hist = model.fit(x_data, y_data, epochs = 500)
```

## 5. 모델 사용

```python
result = model.predict(np.array([-3.5, 3.0, 12.1, 23.1, -26.5]))
result
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Loss trend')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

plt.plot(hist.history['loss'], label='train loss')
plt.legend(loc='best')

plt.show()
```
