```python
import numpy as np
@tf.function
def adder(a,b):
    return a + b

A = tf.constant(1)
B = tf.constant(2)
print(adder(A, B))

C = tf.constant([1,2])
D = tf.constant([3,4])
print(adder(C, D))

E = tf.constant([[1,2,3], [4,5,6]])
F = tf.constant([[2,3,4], [5,6,7]])
print(adder(E, F))
```
